```python
# Documentation site
# https://american-soccer-analysis.github.io/itscalledsoccer/reference/#itscalledsoccer.client.AmericanSoccerAnalysis
from itscalledsoccer.client import AmericanSoccerAnalysis

asa_client = AmericanSoccerAnalysis()
```

```python
# Players salaries
df_players_salary = asa_client.get_player_salaries(leagues='mls', season_name='2025')
df_players_salary
```

```python
# Drop duplicate data and keep latest release date
import pandas as pd
df_players_salary["mlspa_release"] = pd.to_datetime(df_players_salary["mlspa_release"])

df_players_salary = df_players_salary.sort_values("mlspa_release")

df_players_salary = df_players_salary.drop_duplicates(
    subset="player_id",
    keep="last"
)

df_players_salary
```

```python
# Left Join players data set
df_players = asa_client.get_players(leagues='mls')
df_players = df_players[["player_id","player_name","birth_date","nationality","height_ft","height_in","weight_lb"]]

df = df_players_salary.merge(
    df_players,
    on="player_id",
    how="left"
)

df
```

```python
# Left Join teams data set
df_teams = asa_client.get_teams(leagues='mls')

df_teams = df_teams[["team_id", "team_name","team_short_name","team_abbreviation"]]

df_mls = df.merge(
    df_teams,
    on="team_id",
    how="left"
)

df_mls
```

```python
# Get player xG data season 2025
df_xG = asa_client.get_player_xgoals(leagues="mls",season_name='2025')

df_xG = df_xG[["player_id", "general_position","minutes_played","shots","shots_on_target","goals","xgoals","primary_assists","xassists","goals_minus_xgoals","goals_plus_primary_assists","primary_assists_minus_xassists","xgoals_plus_xassists"]]

df_mls25 = df_mls.merge(
    df_xG,
    on="player_id",
    how="left"
)

df_mls25
```

```python
# Check players with missing team
df_mls25["team_id"].isna().sum()
```

```python
# Players with missing team_id values represent cases where the source dataset does not associate the player with a team. 
# This may include players who were waived, retired before or during the season, unsigned free agents, 
# or cases where team data was unavailable or incomplete.  
df_mls25[df_mls25["team_id"].isna()]
```

```python
# Drop rows for players with NaN in team_id
df_mls25 = df_mls25.dropna(subset=["team_id"])
```

```python
# Save dataframe as csv
df_mls25.to_csv("players_mls_25.csv", index=False)
```
